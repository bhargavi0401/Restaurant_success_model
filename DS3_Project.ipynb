```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import json
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
jsonList_bus=[]

with open('yelp_training_set_business.json','r') as file:
    
    for f in file:
        x=json.loads(f)
        jsonList_bus.append(x)
        
```

```python
#Converting data from JSON to CSV
with open('yelp_training_set_business.csv', 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    count = 0
 
    for emp in jsonList_bus:
            if count == 0:
 
        # Writing headers of CSV file
                header = emp.keys()
                csv_writer.writerow(header)
                count += 1
 
    # Writing data of CSV file
            csv_writer.writerow(emp.values())
 
```

```python
std_str=[]
median_str=[]
avg_str=[]
for i in df_bus['business_id']:
    a=df_rev.query(f'business_id == "{i}"')
    median_str.append(a['stars'].median())
    std_str.append(a['stars'].std())
    avg_str.append(a['stars'].mean())
   

```

```python
funny_l=[]
useful_l=[]
cool_l=[]
for i in df_bus['business_id']:
    abc=df_rev.query(f'business_id == "{i}"')
    funny_l.append(abc['funny'].mean())
    useful_l.append(abc['useful'].mean())
    cool_l.append(abc['cool'].mean())
    
```

```python
df_n = df_bus.assign(new_col = median_str)
df_n.rename(columns={"new_col": "median_star"}, inplace=True)
df_n1=df_n.assign(new_col = std_str)
df_n1.rename(columns={"new_col": "std_star"}, inplace=True)
df_new=df_n1.assign(new_col = avg_str)
df_new.rename(columns={"new_col": "mean_star"}, inplace=True)

df_new["funny_m"] = funny_l
df_new["useful_m"] = useful_l
df_new["cool_m"] = cool_l
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Split in features and target
df_ml_features = df_ml_clean.drop('open',axis = 1)
df_ml_target = df_ml_clean['open']
```

```python
# Split the entire dataset
df_train, df_test = train_test_split(df_ml_clean, test_size=0.2, random_state=10)

# Filter open and close businesses
df_open_train = df_train[df_train['open'] == True]
df_close_train = df_train[df_train['open'] == False]
df_open_test = df_test[df_test['open'] == True]
df_close_test = df_test[df_test['open'] == False]
```

```python
len(df_open_train)
len(df_close_train)
```

```python
open_train_class = open_train["open"]
open_test_class = open_test["open"]
close_train_class = close_train["open"]
close_test_class = close_test["open"]
```

```python
# Separate features and target
X_open_train = df_open_train.drop('open', axis=1)
y_open_train = df_open_train['open']
X_close_train = df_close_train.drop('open', axis=1)
y_close_train = df_close_train['open']


X_open_test = df_open_test.drop('open', axis=1)
y_open_test = df_open_test['open']
X_close_test = df_close_test.drop('open', axis=1)
y_close_test = df_close_test['open']
```

```python
# Combine training and testing data for model training
X_train = pd.concat([X_open_train, X_close_train])
y_train = pd.concat([y_open_train, y_close_train])

X_test = pd.concat([X_open_test, X_close_test])
y_test = pd.concat([y_open_test, y_close_test])
```

```python
# Remove 'business_id' column
X_train = X_train.drop('business_id', axis=1)
X_test = X_test.drop('business_id', axis=1)
```

```python


clf = LogisticRegression(random_state=0)
clf_A = GaussianNB()
clf_B = DecisionTreeClassifier(random_state=1)
clf_C = GradientBoostingClassifier(random_state=1)
clf_D = RandomForestClassifier(random_state=1)
```

```python
# Logistic Regression Model
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('Logistic Regression Model:')
print('Accuracy:', clf.score(X_test, y_test))
print('Precision:', precision_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))
print('F1 Score:', f1_score(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
```

```python
# Other Models (clf_A, clf_B, clf_C, clf_D) can be similarly trained and evaluated

# Example for GaussianNB (clf_A)
clf_A = GaussianNB()
clf_A.fit(X_train, y_train)
y_pred_A = clf_A.predict(X_test)
print('\nGaussianNB Model:')
print('Accuracy:', clf_A.score(X_test, y_test))
print('Precision:', precision_score(y_test, y_pred_A))
print('Recall:', recall_score(y_test, y_pred_A))
print('F1 Score:', f1_score(y_test, y_pred_A))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_A))
```

```python
# Decision Tree Classifier (clf_B)
clf_B = DecisionTreeClassifier(random_state=1)
clf_B.fit(X_train, y_train)
y_pred_B = clf_B.predict(X_test)

print('\nDecision Tree Classifier (clf_B):')
print('Accuracy:', clf_B.score(X_test, y_test))
print('Precision:', precision_score(y_test, y_pred_B))
print('Recall:', recall_score(y_test, y_pred_B))
print('F1 Score:', f1_score(y_test, y_pred_B))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_B))
```

```python
# Gradient Boosting Classifier (clf_C)
clf_C = GradientBoostingClassifier(random_state=1)
clf_C.fit(X_train, y_train)
y_pred_C = clf_C.predict(X_test)

print('\nGradient Boosting Classifier (clf_C):')
print('Accuracy:', clf_C.score(X_test, y_test))
print('Precision:', precision_score(y_test, y_pred_C))
print('Recall:', recall_score(y_test, y_pred_C))
print('F1 Score:', f1_score(y_test, y_pred_C))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_C))
```

```python
# Random Forest Classifier (clf_D)
clf_D = RandomForestClassifier(random_state=1)
clf_D.fit(X_train, y_train)
y_pred_D = clf_D.predict(X_test)

print('\nRandom Forest Classifier (clf_D):')
print('Accuracy:', clf_D.score(X_test, y_test))
print('Precision:', precision_score(y_test, y_pred_D))
print('Recall:', recall_score(y_test, y_pred_D))
print('F1 Score:', f1_score(y_test, y_pred_D))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred_D))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Get decision functions or predicted probabilities for each classifier
y_score = clf.decision_function(X_test)
y_score_A = clf_A.predict_proba(X_test)[:, 1]
y_score_B = clf_B.predict_proba(X_test)[:, 1]
y_score_C = clf_C.decision_function(X_test)
y_score_D = clf_D.predict_proba(X_test)[:, 1]

# Compute ROC curve and ROC area for each classifier
fpr, tpr, _ = roc_curve(y_test.values, y_score)
roc_auc = auc(fpr, tpr)

fpr_A, tpr_A, _ = roc_curve(y_test.values, y_score_A)
roc_auc_A = auc(fpr_A, tpr_A)

fpr_B, tpr_B, _ = roc_curve(y_test.values, y_score_B)
roc_auc_B = auc(fpr_B, tpr_B)

fpr_C, tpr_C, _ = roc_curve(y_test.values, y_score_C)
roc_auc_C = auc(fpr_C, tpr_C)

fpr_D, tpr_D, _ = roc_curve(y_test.values, y_score_D)
roc_auc_D = auc(fpr_D, tpr_D)

# Plot ROC curves
plt.figure(figsize=(8, 6))
plt.plot([0, 1], [0, 1], 'k--', label='Random')

plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {roc_auc:.2f})')
plt.plot(fpr_A, tpr_A, label=f'Gaussian NB (AUC = {roc_auc_A:.2f})')
plt.plot(fpr_B, tpr_B, label=f'Decision Tree (AUC = {roc_auc_B:.2f})')
plt.plot(fpr_C, tpr_C, label=f'Gradient Boosting (AUC = {roc_auc_C:.2f})')
plt.plot(fpr_D, tpr_D, label=f'Random Forest (AUC = {roc_auc_D:.2f})')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
